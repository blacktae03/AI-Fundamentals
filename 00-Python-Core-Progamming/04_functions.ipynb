## AI, 파이썬, 그리고 자연로그: 왜 우리는 `e`와 `ln`을 사용할까?

AI와 머신러닝 코드를 다루다 보면 `np.log()`나 `math.log()`처럼 로그를 사용하는 경우를 자주 마주하게 됩니다. AI 모델이 학습하는 핵심 원리인 **미분(Calculus)의 편리함**에 있습니다.

### 숫자 $e$ 와 자연로그 $\ln$

1.  **숫자 $e$ (오일러 수, Euler's number)**
    *   **정의**: 숫자 **$e$** 는 약 2.71828... 값을 가지는 무리수입니다. 이 상수는 자연계의 모든 **'연속적인 성장과 변화'**를 설명하는 가장 기본적인 값입니다.

    가장 대표적인 예가 바로 **'복리 이자'**의 극한 개념입니다.
    연이율 100%의 이자를 주는 은행이 있다고 상상해 봅시다. 이 이자를 주는 횟수를 무한히 잘게 쪼개어 '모든 순간마다' 이자가 원금에 더해지는 **연속 복리** 상황이 될 때, 1년 뒤 원금은 최종적으로 약 2.718배가 됩니다.

    이를 수학적으로는 다음과 같이 표현합니다.

    $$ e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n \approx 2.71828... $$

    예를 들어, 100% 이자를 '매일' 단위($n=365$)로 잘게 쪼개서 복리로 계산하면 그 값은 $e$에 매우 가까워집니다.

    $$ \left(1 + \frac{1}{365}\right)^{365} \approx 2.7145 $$
    *   **AI에서의 역할**: 주로 **지수 함수 $e^x$** 형태로 등장하며,
    
    시그모이드(Sigmoid) $$\sigma(x) = \frac{1}{1 + e^{-x}}$$ 소프트맥스(Softmax) $$\text{Softmax}(z_j) = \frac{e^{z_j}}{\sum_{k=1}^{K} e^{z_k}} \quad \text{for } j = 1, \dots, K$$ 같은 함수에 사용됩니다.

2.  **자연로그 $\ln(x)$ (Natural Logarithm, $\log_e(x)$)**
    *   **정의**: "숫자 $e$ 를 몇 번 제곱해야 $x$ 가 될까?"라는 질문에 대한 답입니다. 즉, 지수 함수 $e^x$ 의 역함수 관계입니다. ($ \ln(e^x) = x $)
    *   **파이썬에서의 함정**: **`math.log(x)`와 `np.log(x)`는 모두 밑이 $e$ 인 자연로그입니다.** 밑이 10인 로그를 쓰고 싶다면 `math.log10(x)`를 사용해야 합니다.

### AI에서 자연로그를 사용하는 3가지 핵심 이유

#### 1. 미분이 단순하다

AI 모델 학습은 손실 함수(Loss function)를 최소화하기 위해 수백만 개의 파라미터를 미분하는 과정의 연속입니다. 이때 자연로그의 미분 결과는 단순합니다.

*   **자연로그의 미분**
    $$ \frac{d}{dx} \ln(x) = \frac{1}{x} $$
    결과가 깔끔하여 수학적으로 다루기 편리합니다.


#### 2. 정보 이론과의 조화

AI 분류 문제의 핵심 손실 함수인 **크로스 엔트로피(Cross-Entropy)**는 정보 이론에 기반합니다. 정보 이론에서 정보량을 측정하는 '자연스러운' 단위(nat)는 바로 자연로그를 사용할 때 정의됩니다.

*   **엔트로피 공식**:
    $$ H(P) = -\sum_{i} P(i) \log(P(i)) $$
    이 공식에서 자연로그를 사용하는 것은 연속적인 확률 분포를 다루는 데 수학적으로 일관되고 편리하여 AI 분야에서 많이 사용됩니다.

#### 3. 지수 함수($e^x$) 의 역함수

AI 모델의 출력값을 확률로 변환해주는 소프트맥스(Softmax) 함수는 $e^x$를 기반으로 합니다.

*   **소프트맥스 함수**:
    $$ \text{Softmax}(z_j) = \frac{e^{z_j}}{\sum_{k} e^{z_k}} $$
    이 출력값에 대한 손실을 계산하기 위해 자연로그를 씌우면 $ \ln(e^{\dots}) $ 형태가 되어 수식이 매우 간단하게 정리됩니다.
    $$\ln(e^{z_j}) - \ln \left( \sum_{k} e^{z_k} \right)$$
    $$ z_j - \ln \left( \sum_{k} e^{z_k} \right) $$
    이처럼 $e$ 기반의 함수들과 자연스럽게 어우러지기 때문에 자연로그가 함께 사용됩입니다.


## 실습 문제 7: 정규분포 확률 밀도 함수(PDF) 구현

**설명**: NumPy와 같은 라이브러리는 정규분포 샘플링을 위한 편리한 함수를 제공하지만, 그 근간이 되는 확률 밀도 함수(PDF, Probability Density Function)를 직접 구현해보는 것은 개념을 이해하는 데 큰 도움이 됩니다. 이 실습에서는 정규분포의 PDF 공식을 코드로 직접 작성하고, 이 함수의 그래프가 실제 데이터 샘플의 분포와 일치하는지 시각적으로 확인합니다.

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\!\Bigl(-\frac{(x-\mu)^2}{2\sigma^2}\Bigr)$$

  - $\\mu$ : 평균 (mean)
  - $ \\sigma\\ $: 표준편차 (standard deviation)

**요구사항**:

  - 주어진 정규분포 PDF 공식을 파이썬 코드로 변환하여 `f(x, mu, sigma)` 함수를 완성하세요. `np.pi`와 `np.exp`를 활용하면 편리합니다.
  - 평균(`μ`) 0, 표준편차(`σ`) 1인 표준 정규분포의 PDF를 계산하도록 함수를 호출하세요.
  - `np.random.normal`로 생성한 샘플 데이터의 히스토그램과 직접 구현한 `f(x)` 함수의 그래프를 함께 그려, 두 결과가 일치하는지 확인합니다.

```python
import numpy as np

a = np.array([1.0, 2.0, 3.0])
a *= np.pi
b = np.sqrt(np.pi)
c = np.array([1.0 ,2.0 ,3.0])
c *= b

print(a)
print(b)
print(type(b))
b *= 3
print(type(b))
print(c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x: np.ndarray, mu: float, sigma: float) -> np.ndarray:
  """
  정규분포의 확률 밀도 함수(PDF)를 계산합니다.
  x: 확률 변수 값(들)
  mu: 평균
  sigma: 표준편차
  """
  # TODO: 정규분포 PDF 공식을 코드로 구현하세요.
  return (1 / (sigma * np.sqrt(2*np.pi))) * np.exp(-1 * (x-mu)*(x-mu) / 2*sigma*sigma)
  pass


mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
# print(type(x))
y = f(x, mu, sigma)

plt.plot(x, y)
```
