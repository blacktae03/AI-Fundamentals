## ì‹¤ìŠµ ë¬¸ì œ: ì„­ì”¨-í™”ì”¨ ì–‘ë°©í–¥ ì˜¨ë„ ë³€í™˜ê¸° ì œì‘ ë° í…ŒìŠ¤íŠ¸

**ì„¤ëª…**: ì´ì „ ì‹¤ìŠµì„ í™•ì¥í•˜ì—¬, ì´ë²ˆì—ëŠ” ì„­ì”¨ë¥¼ í™”ì”¨ë¡œ, ê·¸ë¦¬ê³  í™”ì”¨ë¥¼ ì„­ì”¨ë¡œ ë³€í™˜í•˜ëŠ” ë‘ ê°œì˜ ê°œë³„ í•¨ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤. í•¨ìˆ˜ë¥¼ ê¸°ëŠ¥ ë‹¨ìœ„ë¡œ ë¶„ë¦¬í•˜ëŠ” ì—°ìŠµì„ í•˜ê³ , ê° í•¨ìˆ˜ê°€ ì˜¬ë°”ë¥´ê²Œ ì‘ë™í•˜ëŠ”ì§€ í™•ì¸í•˜ê¸° ìœ„í•œ ê°„ë‹¨í•œ í…ŒìŠ¤íŠ¸ ì½”ë“œë¥¼ ì‘ì„±í•˜ì—¬ ëª¨ë“ˆí™”ëœ í”„ë¡œê·¸ë˜ë°ì˜ ê¸°ì´ˆë¥¼ í•™ìŠµí•©ë‹ˆë‹¤.

$$F = C \times \frac{9}{5} + 32 \quad \text{ê·¸ë¦¬ê³ } \quad C = (F - 32) \times \frac{5}{9}$$

**ìš”êµ¬ì‚¬í•­**:

  - **í•¨ìˆ˜ 1**: `celsius_to_fahrenheit` í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì—¬ ì„­ì”¨ ì˜¨ë„ë¥¼ í™”ì”¨ ì˜¨ë„ë¡œ ë³€í™˜í•˜ì„¸ìš”.
  - **í•¨ìˆ˜ 2**: `fahrenheit_to_celsius` í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì—¬ í™”ì”¨ ì˜¨ë„ë¥¼ ì„­ì”¨ ì˜¨ë„ë¡œ ë³€í™˜í•˜ì„¸ìš”.
  - **í…ŒìŠ¤íŠ¸ ì½”ë“œ**: ì œê³µëœ í…ŒìŠ¤íŠ¸ ì½”ë“œ ì„¹ì…˜ì—ì„œ ê° í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ë³€í™˜ì´ ì •í™•í•œì§€ í™•ì¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ì„¸ìš”. (ì˜ˆ: 0Â°C â†’ 32Â°F, 212Â°F â†’ 100Â°C)



```python

def celsius_to_fahrenheit(celsius: float) -> float:
    """ì„­ì”¨(Celsius)ë¥¼ í™”ì”¨(Fahrenheit)ë¡œ ë³€í™˜í•©ë‹ˆë‹¤."""
    # ì´ê³³ì— ì„­ì”¨ -> í™”ì”¨ ë³€í™˜ ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
    return celsius * 9 / 5 + 32
    pass

def fahrenheit_to_celsius(fahrenheit: float) -> float:
    """í™”ì”¨(Fahrenheit)ë¥¼ ì„­ì”¨(Celsius)ë¡œ ë³€í™˜í•©ë‹ˆë‹¤."""
    # ì´ê³³ì— í™”ì”¨ -> ì„­ì”¨ ë³€í™˜ ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
    return (fahrenheit-32) * 5 / 9
    pass
```

```python
# í…ŒìŠ¤íŠ¸ 1: 0Â°Cê°€ 32Â°Fë¡œ ë³€í™˜ë˜ëŠ”ì§€ í™•ì¸
c_temp = 0
f_expected = 32
f_actual = celsius_to_fahrenheit(c_temp)
print(f"{c_temp}Â°C -> {f_actual}Â°F (ì˜ˆìƒ: {f_expected}Â°F)")
print(f"í…ŒìŠ¤íŠ¸ í†µê³¼: {f_actual == f_expected}")

print("-" * 20)

# í…ŒìŠ¤íŠ¸ 2: 212Â°Fê°€ 100Â°Cë¡œ ë³€í™˜ë˜ëŠ”ì§€ í™•ì¸
f_temp = 212
c_expected = 100
c_actual = fahrenheit_to_celsius(f_temp)
print(f"{f_temp}Â°F -> {c_actual}Â°C (ì˜ˆìƒ: {c_expected}Â°C)")
print(f"í…ŒìŠ¤íŠ¸ í†µê³¼: {c_actual == c_expected}")
```

## ì‹¤ìŠµ ë¬¸ì œ: ë¶€ë™ì†Œìˆ˜ì ì˜ ì•ˆì „í•œ ë¹„êµ í•¨ìˆ˜ êµ¬í˜„í•˜ê¸°
**ì„¤ëª…**: ì»´í“¨í„°ëŠ” ì‹­ì§„ìˆ˜ ì†Œìˆ˜ë¥¼ ì´ì§„ìˆ˜ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì—ì„œ ë¯¸ì„¸í•œ ì˜¤ì°¨ë¥¼ ê°€ì§‘ë‹ˆë‹¤. ì´ë¡œ ì¸í•´ `0.1 + 0.2`ëŠ” ì •í™•íˆ `0.3`ì´ ì•„ë‹ˆë¯€ë¡œ `==` ì—°ì‚°ìë¡œ ë¹„êµí•˜ë©´ `False`ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì´ ë¬¸ì œë¥¼ í•´ê²°í•˜ê¸° ìœ„í•´, ë‘ ìˆ«ìì˜ ì°¨ì´ê°€ ì•„ì£¼ ì‘ì€ ê°’(í—ˆìš© ì˜¤ì°¨, ì—¡ì‹¤ë¡ )ë³´ë‹¤ ì‘ì€ì§€ë¥¼ í™•ì¸í•˜ëŠ” ë°©ë²•ì´ ì‚¬ìš©ë©ë‹ˆë‹¤. ë‘ ë¶€ë™ì†Œìˆ˜ì  ìˆ«ìë¥¼ ì•ˆì „í•˜ê²Œ ë¹„êµí•˜ëŠ” í•¨ìˆ˜ë¥¼ ì§ì ‘ êµ¬í˜„í•´ ë³´ì„¸ìš”.

$$ |a - b| < \epsilon $$

**ìš”êµ¬ì‚¬í•­**:
- `are_floats_equal`ì´ë¼ëŠ” ì´ë¦„ì˜ í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì—¬, ë‘ ë¶€ë™ì†Œìˆ˜ì  `a`ì™€ `b`ì˜ ì°¨ì´(ì ˆëŒ“ê°’)ê°€ ì£¼ì–´ì§„ í—ˆìš© ì˜¤ì°¨ `epsilon`ë³´ë‹¤ ì‘ì€ì§€ ê²€ì‚¬í•˜ì„¸ìš”.
- `a`ì— `0.1 + 0.2`ë¥¼, `b`ì— `0.3`ì„ í• ë‹¹í•˜ì„¸ìš”.
- ì™„ì„±ëœ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ `a`ì™€ `b`ê°€ ì‹¤ì§ˆì ìœ¼ë¡œ ê°™ì€ì§€ í™•ì¸í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ì„¸ìš”.



```python
def are_floats_equal(a: float, b: float, epsilon: float = 1e-9) -> bool:
    """
    ë‘ ë¶€ë™ì†Œìˆ˜ì  ìˆ«ìê°€ í—ˆìš© ì˜¤ì°¨(epsilon) ë‚´ì—ì„œ ê°™ì€ì§€ í™•ì¸í•©ë‹ˆë‹¤.

    Args:
        a (float): ì²« ë²ˆì§¸ ë¶€ë™ì†Œìˆ˜ì  ìˆ«ì.
        b (float): ë‘ ë²ˆì§¸ ë¶€ë™ì†Œìˆ˜ì  ìˆ«ì.
        epsilon (float): í—ˆìš©ë˜ëŠ” ìµœëŒ€ ì˜¤ì°¨. ê¸°ë³¸ê°’ì€ 1e-9ì…ë‹ˆë‹¤.

    Returns:
        bool: ë‘ ìˆ«ìê°€ í—ˆìš© ì˜¤ì°¨ ë‚´ì—ì„œ ê°™ìœ¼ë©´ True, ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ False.
    """
    # return a - b < epsilon or a - b > -epsilon
    return abs(a-b) < epsilon
    # ì´ ë¶€ë¶„ì„ êµ¬í˜„í•˜ì„¸ìš”.
    pass
```

```python
# í…ŒìŠ¤íŠ¸ ì½”ë“œ
num1 = 0.1 + 0.2
num2 = 0.3

# í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ê²°ê³¼ë¥¼ í™•ì¸í•˜ê³  ì¶œë ¥í•˜ì„¸ìš”.
result = are_floats_equal(num1, num2)
print(f"0.1 + 0.2ì™€ 0.3ì€ ê°™ì€ê°€? {result}")
```

## ì‹¤ìŠµ ë¬¸ì œ: ë¶€ë™ì†Œìˆ˜ì  ëˆ„ì  ì˜¤ì°¨ ì‹œë®¬ë ˆì´ì…˜
**ì„¤ëª…**: ë‹¨ì¼ ë¶€ë™ì†Œìˆ˜ì  ì—°ì‚°ì˜ ì˜¤ì°¨ëŠ” ë§¤ìš° ì‘ì§€ë§Œ, ì´ëŸ¬í•œ ì‘ì€ ì˜¤ì°¨ë“¤ì´ ë°˜ë³µë¬¸ ì•ˆì—ì„œ ìˆ˜ì²œ, ìˆ˜ë§Œ ë²ˆ ë”í•´ì§€ë©´ ë¬´ì‹œí•  ìˆ˜ ì—†ëŠ” **ëˆ„ì  ì˜¤ì°¨(Accumulated Error)** ë¡œ ì»¤ì§ˆ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ” `for` ë£¨í”„ë¥¼ ì‚¬ìš©í•´ ì‘ì€ ì†Œìˆ˜(`0.1`)ë¥¼ ì—¬ëŸ¬ ë²ˆ ë”í•˜ëŠ” ê°„ë‹¨í•œ ì‹œë®¬ë ˆì´ì…˜ì„ í†µí•´, ì´ë¡ ì ì¸ ê²°ê³¼ì™€ ì‹¤ì œ ê³„ì‚° ê²°ê³¼ ì‚¬ì´ì— ì˜¤ì°¨ê°€ ì–´ë–»ê²Œ ë°œìƒí•˜ëŠ”ì§€ ì§ì ‘ í™•ì¸í•´ ë³´ê² ìŠµë‹ˆë‹¤. ğŸ’»

$$ \sum_{i=1}^{N} x \neq N \times x \quad (\text{in floating-point arithmetic}) $$

**ìš”êµ¬ì‚¬í•­**:
- `accumulate_error` í•¨ìˆ˜ ë‚´ì—ì„œ `for` ë£¨í”„ë¥¼ ì‚¬ìš©í•´ `increment` ê°’ì„ `iterations` íšŸìˆ˜ë§Œí¼ `total`ì— ë”í•˜ëŠ” ì½”ë“œë¥¼ ì™„ì„±í•˜ì„¸ìš”.
- `0.1`ì„ `10000`ë²ˆ ë”í•œ ê²°ê³¼ë¥¼ í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ì–»ìœ¼ì„¸ìš”.
- ìˆ˜í•™ì ìœ¼ë¡œ ì˜ˆìƒë˜ëŠ” ê²°ê³¼(`0.1 * 10000`)ì™€ ë£¨í”„ë¥¼ í†µí•´ ì–»ì€ ì‹¤ì œ ê²°ê³¼ë¥¼ ë¹„êµí•˜ê³ , ê·¸ ì°¨ì´(ëˆ„ì  ì˜¤ì°¨)ë¥¼ ì¶œë ¥í•˜ì—¬ í™•ì¸í•˜ì„¸ìš”.


```python

def accumulate_error(increment: float, iterations: int) -> float:
    """
    ì£¼ì–´ì§„ increment ê°’ì„ iterations íšŸìˆ˜ë§Œí¼ ë°˜ë³µí•´ì„œ ë”í•˜ì—¬
    ë¶€ë™ì†Œìˆ˜ì ì˜ ëˆ„ì  ì˜¤ì°¨ë¥¼ í™•ì¸í•©ë‹ˆë‹¤.

    Args:
        increment (float): ê° ë°˜ë³µë§ˆë‹¤ ë”í•  ê°’.
        iterations (int): ë°˜ë³µ íšŸìˆ˜.

    Returns:
        float: ëˆ„ì ëœ í•©ê³„.
    """
    total = 0.0
    # for ë£¨í”„ë¥¼ ì‚¬ìš©í•˜ì—¬ incrementë¥¼ totalì— iterations ë§Œí¼ ë”í•˜ì„¸ìš”.
    # ì´ ë¶€ë¶„ì„ êµ¬í˜„í•˜ì„¸ìš”.
    for _ in range(iterations) :
        total += increment

    return total

```

```python
# ì‹œë®¬ë ˆì´ì…˜ íŒŒë¼ë¯¸í„°
inc = 0.1
iters = 10000

# í•¨ìˆ˜ í˜¸ì¶œ ë° ê²°ê³¼ ê³„ì‚°
actual_sum = accumulate_error(inc, iters)
expected_sum = inc * iters
error = actual_sum - expected_sum

print(f"ë£¨í”„ë¥¼ ì´ìš©í•œ ê³„ì‚° ê²°ê³¼: {actual_sum}")
print(f"ìˆ˜í•™ì  ì˜ˆìƒ ê²°ê³¼: {expected_sum}")
print(f"ëˆ„ì  ì˜¤ì°¨: {error}")
```

## ì‹¤ìŠµ ë¬¸ì œ: ë°˜ë³µë²•ì„ ì´ìš©í•œ ì œê³±ê·¼ ê·¼ì‚¬ê°’ ê³„ì‚°
**ì„¤ëª…**: ì–´ë–¤ ê°’ì— ì ì°¨ ê°€ê¹Œì›Œì§€ëŠ” ê³„ì‚°ì„ ë°˜ë³µí•˜ëŠ” ê²ƒì„ **ë°˜ë³µë²•(Iterative Method)** ì´ë¼ê³  í•©ë‹ˆë‹¤. ì´ëŠ” ìˆ˜ì¹˜ í•´ì„ì˜ í•µì‹¬ ì›ë¦¬ ì¤‘ í•˜ë‚˜ì…ë‹ˆë‹¤. ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ” <a href="https://en.wikipedia.org/wiki/Square_root_algorithms#Heron's_method">í—¤ë¡ ì˜ ë°©ë²•</a>ì„ ì´ìš©í•˜ì—¬ ìˆ«ìì˜ ì œê³±ê·¼ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤. ì´ ë°©ë²•ì€ ì¶”ì¸¡ê°’($x_n$)ì„ ê³„ì†í•´ì„œ ë” ì •í™•í•œ ê°’($x_{n+1}$)ìœ¼ë¡œ ê°±ì‹ í•˜ë©° ì‹¤ì œ ì œê³±ê·¼ì— ìˆ˜ë ´ì‹œí‚µë‹ˆë‹¤. `while` ë£¨í”„ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‘ ê°’ì˜ ì°¨ì´ê°€ ì•„ì£¼ ì‘ì•„ì§ˆ ë•Œê¹Œì§€ ì´ ê³¼ì •ì„ ë°˜ë³µí•´ ë³´ì„¸ìš”.

$$ x_{n+1} = \frac{1}{2} \left( x_n + \frac{S}{x_n} \right) $$

**ìš”êµ¬ì‚¬í•­**:
- `sqrt_iterative` í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì„¸ìš”. ì´ í•¨ìˆ˜ëŠ” ì œê³±ê·¼ì„ êµ¬í•  ìˆ«ì `S`ì™€ ì´ˆê¸° ì¶”ì¸¡ê°’ `guess`ë¥¼ ì¸ìë¡œ ë°›ìŠµë‹ˆë‹¤.
- `while` ë£¨í”„ë¥¼ ì‚¬ìš©í•˜ì—¬ ì¶”ì¸¡ê°’ì„ ê³„ì† ê°±ì‹ í•˜ì„¸ìš”. ë£¨í”„ëŠ” ì´ì „ ì¶”ì¸¡ê°’ê³¼ ìƒˆë¡œìš´ ì¶”ì¸¡ê°’ì˜ ì°¨ì´ê°€ í—ˆìš© ì˜¤ì°¨(`1e-9`)ë³´ë‹¤ ì‘ì•„ì§€ë©´ ë©ˆì¶°ì•¼ í•©ë‹ˆë‹¤.
- í•¨ìˆ˜ê°€ ìµœì¢…ì ìœ¼ë¡œ ê³„ì‚°ëœ ì œê³±ê·¼ì˜ ê·¼ì‚¬ê°’ì„ ë°˜í™˜í•˜ë„ë¡ í•˜ì„¸ìš”.
- `sqrt_iterative(2)`ë¥¼ í˜¸ì¶œí•˜ì—¬ ê²°ê³¼ë¥¼ í™•ì¸í•˜ê³ , `math.sqrt(2)`ì˜ ê²°ê³¼ì™€ ë¹„êµí•´ ë³´ì„¸ìš”.


```python
import math

def sqrt_iterative(S: float, guess: float = 1.0) -> float:
    """
    í—¤ë¡ ì˜ ë°©ë²•ì„ ì‚¬ìš©í•˜ì—¬ ìˆ«ì Sì˜ ì œê³±ê·¼ì„ ë°˜ë³µì ìœ¼ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤.

    Args:
        S (float): ì œê³±ê·¼ì„ ì°¾ê³ ì í•˜ëŠ” ìˆ«ì.
        guess (float): ì´ˆê¸° ì¶”ì¸¡ê°’.

    Returns:
        float: Sì˜ ì œê³±ê·¼ì— ëŒ€í•œ ê·¼ì‚¬ê°’.
    """
    # í—ˆìš© ì˜¤ì°¨
    tolerance = 1e-9
    # while True :
    #     guessBack = guess
    #     guess = (guess + S / guess) / 2
    #     if guess - guessBack < tolerance: break

    while abs(guess**2 - S) >= tolerance :
        guess = (guess + S/guess) / 2

    # ì´ ë¶€ë¶„ì„ êµ¬í˜„í•˜ì„¸ìš”.

    return guess
```

```python
# í…ŒìŠ¤íŠ¸ ì½”ë“œ
number_to_sqrt = 2
approximated_sqrt = sqrt_iterative(number_to_sqrt)
actual_sqrt = math.sqrt(number_to_sqrt)

print(f"ë°˜ë³µë²•ìœ¼ë¡œ ê³„ì‚°í•œ {number_to_sqrt}ì˜ ì œê³±ê·¼: {approximated_sqrt}")
print(f"math ëª¨ë“ˆë¡œ ê³„ì‚°í•œ {number_to_sqrt}ì˜ ì œê³±ê·¼: {actual_sqrt}")
```
